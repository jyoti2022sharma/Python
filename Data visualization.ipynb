```python
df = pd.read_csv("sharma-kohli.csv")
df.head(2)
```

```python
plt.plot(df["index"], df["V Kohli"])
```

```python
# if you want to see two plot together 
plt.plot(df["index"], df["V Kohli"])
plt.plot(df["index"], df["RG Sharma"])
```

# labels title

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(df['index'],df['V Kohli'],color='green')
plt.plot(df['index'],df['RG Sharma'],color='blue')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# if instaed od solid line, you need dash line
plt.plot(df['index'],df['V Kohli'],color='green', linestyle = "dashed")
plt.plot(df['index'],df['RG Sharma'],color='blue')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

 marker parameter 
https://matplotlib.org/stable/api/markers_api.html


```python
# if instaed od solid line, you need dash line
plt.plot(df['index'],df['V Kohli'],color='green', linestyle = "dashed", linewidth = 3, marker = "D", markersize = 10)
plt.plot(df['index'],df['RG Sharma'],color='blue')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(df['index'],df['V Kohli'],color='green', linestyle = "dashed", linewidth = 3, marker = "D", markersize = 10)
plt.plot(df['index'],df['RG Sharma'],color='blue')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='upper right')

plt.show
```

# plt show

this plt.show is main line that we need to show to at the end, in some text editor if we do not show this, the graph will not display

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grid
plt.plot(df['index'],df['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(df['index'],df['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```
